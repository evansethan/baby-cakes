```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Your matrix
data = {
    'profanity':       [1.000000, 0.119168, 0.521290, -0.015528],
    'subjectivity':    [0.119168, 1.000000, 0.084485, -0.001354],
    'toxicity':        [0.521290, 0.084485, 1.000000, 0.011656],
    'insider_terms':   [-0.015528, -0.001354, 0.011656, 1.000000]
}
matrix = pd.DataFrame(data, index=['profanity', 'subjectivity', 'toxicity', 'insider_terms'])

plt.figure(figsize=(6, 5))
sns.heatmap(matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar=True, square=True)
plt.title("Feature Correlation Matrix")
plt.tight_layout()
plt.show()
```

```python
# Data from classification report
data = {
    'class': ['non_extreme', 'extreme'],
    'precision': [0.94, 0.93],
    'recall': [0.97, 0.88],
    'f1-score': [0.95, 0.90]
}

df = pd.DataFrame(data)
df.set_index('class', inplace=True)

# Plot
df.plot(kind='bar', figsize=(8, 5))
plt.title("Model Evaluation (LogReg)")
plt.ylabel("Score")
plt.ylim(0.8, 1.0)
plt.xticks(rotation=0)
plt.legend(loc='lower right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
